```python
# Take any random samples
import numpy as np
import matplotlib.pyplot as plt

# Parameters
data_size = 1000  # Size of the dataset
skewness = 2     # Adjust this value to control skewness (higher values create stronger skew)

# Generate right-skewed data
right_skewed_data = np.random.exponential(scale=skewness, size=data_size)

# Plot the data
plt.hist(right_skewed_data, bins=30, density=True, alpha=0.6, color='b')
plt.title("Right-Skewed Data")
plt.show()
```

```python
mean = np.mean(right_skewed_data)
mean


# std_dev = np.std(right_skewed_data)
# std_dev
```

```python
population_size = 1000
sample_size = 30
num_samples = 500

sample_means = []
for _ in range(num_samples):
  sample = np.random.choice(right_skewed_data, size=30, replace=True)
  sample_mean = np.mean(sample)
  sample_means.append(sample_mean)
```

```python
# Plot the histogram of sample means
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='b')
plt.title("Distribution of Sample Means")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

```python
mean = np.mean(sample_means)
mean
```

```python
std_dev = np.std(sample_means)
std_dev
```
